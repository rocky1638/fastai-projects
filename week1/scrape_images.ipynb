# Dog or Fried Chicken?
This is my first project while following along with fast.ai, where I attempt to train a CNN to classify between two classes, dogs and fried chicken.

All this first part does is scrape the images required for the training from Google Images.

```python
from google_images_download import google_images_download
```

Instantiate the Google Image Downloader

```python
response = google_images_download.googleimagesdownload()
```

Download 300 pictures of curly haired dogs to `./downloads/dog`

```python
arguments = {
    "keywords": "curly hair dog",
    "limit": 300,
    "print_urls": True,
    "color_type": "full-color",
    "image_directory": "dog",
    "chromedriver": "/usr/bin/chromedriver"
}

absolute_image_paths = response.download(arguments)
```

Download 300 pictures of fried chicken to `./downloads/fried_chicken`

```python
arguments = {
    "keywords": "fried chicken",
    "limit": 300,
    "print_urls": False,
    "color_type": "full-color",
    "image_directory": "fried_chicken",
    "chromedriver": "/usr/bin/chromedriver"
}

absolute_image_paths = response.download(arguments)
```
